# Kesikli Olasılık Dağılımları:

# 1-Bernoulli Dağılımı:

Bernoulli Dağılımı:

- Bir deneyde sadece iki olası sonuç (başarılı/başarısız, evet/hayır) bulunan deneyler için kullanılır.
- Bu dağılım, bir denemenin başarı olasılığını (p) ve başarısızlık olasılığını (q = 1 - p) temsil eder.
- Tek bir deneme sonucu için olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=x) = p^x * q^(1-x) (x=0,1)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Başarı olasılığı
p = 0.3

# X değerleri (başarılı: 1, başarısız: 0)
x = [0, 1]

# Bernoulli Dağılımı olasılık kütle fonksiyonu (PMF)
pmf = bernoulli.pmf(x, p)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x, ['Başarısız', 'Başarılı'])
plt.ylabel('Olasılık')
plt.xlabel('Sonuçlar')
plt.title('Bernoulli Dağılımı')
plt.show()
```

# 2-Binom Dağılımı:

Binom Dağılımı:

- Bağımsız n deneme sonucunda r başarı sayısını modellemek için kullanılır.
- Bu dağılım, belirli bir deneme sayısı n, başarı olasılığı p ve başarısızlık olasılığı q = 1 - p ile karakterizedir.
- Olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=r) = C(n, r) * p^r * q^(n-r)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Deneme sayısı
n = 10

# Başarı olasılığı
p = 0.5

# X değerleri (başarı sayısı)
x = np.arange(0, n+1)

# Binom Dağılımı olasılık kütle fonksiyonu (PMF)
pmf = binom.pmf(x, n, p)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x)
plt.ylabel('Olasılık')
plt.xlabel('Başarı Sayısı')
plt.title('Binom Dağılımı (n=10, p=0.5)')
plt.show()
```

# 3-Poisson Dağılımı:


Poisson Dağılımı:

* Belirli bir zaman aralığında nadir olarak gerçekleşen olayların sayısını modellemek için kullanılır.
* Poisson Dağılımı, sürekli zaman aralıkları ve düşük olasılıklı nadir olaylarla ilgilenir.
* Olayların ortalama frekansını λ ile ifade eder.
* Olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=k) = (e^(-λ) * λ^k) / k!




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Olay sayısı ortalama değeri (lambda)
lam = 3

# X değerleri (olay sayısı)
x = np.arange(0, 10)

# Poisson Dağılımı olasılık kütle fonksiyonu (PMF)
pmf = poisson.pmf(x, lam)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x)
plt.ylabel('Olasılık')
plt.xlabel('Olay Sayısı')
plt.title('Poisson Dağılımı (lambda=3)')
plt.show()
```

# 4-Geometrik Dağılım:

Geometrik Dağılım:

* Bağımsız denemelerde başarı olasılığı p olan bir olayın ilk başarıya ulaşma denemesinin olasılığını modellemek için kullanılır.
* Olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=k) = (1 - p)^(k-1) * p (k=1,2,3,...)




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Başarı olasılığı
p = 0.2

# X değerleri (deneme sayısı)
x = np.arange(1, 11)

# Geometrik Dağılım olasılık kütle fonksiyonu (PMF)
pmf = geom.pmf(x, p)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x)
plt.ylabel('Olasılık')
plt.xlabel('Deneme Sayısı')
plt.title('Geometrik Dağılım (p=0.2)')
plt.show()
```

# 5-Hipergeometrik Dağılım:


Hipergeometrik Dağılım:

* Rastgele örneklemelerde, popülasyondan çekilen belirli örneklem sayılarıyla ilgili problemleri modellemek için kullanılır.
* N: Toplam popülasyon, K: İyi durumdaki toplam öğe sayısı, n: Çekilecek örneklem sayısı.
* Olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=k) = (C(K, k) * C(N-K, n-k)) / C(N, n) (k=min(K,n), min(K,n)+1, ..., max(0,n-K)+min(K,n))




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Toplam popülasyon
N = 50

# İyi durumdaki toplam öğe sayısı
K = 10

# Çekilecek örneklem sayısı
n = 5

# X değerleri (başarı sayısı)
x = np.arange(max(0, n + K - N), min(n, K) + 1)

# Hipergeometrik Dağılım olasılık kütle fonksiyonu (PMF)
pmf = hypergeom.pmf(x, N, K, n)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x)
plt.ylabel('Olasılık')
plt.xlabel('Başarı Sayısı')
plt.title('Hipergeometrik Dağılım (N=50, K=10, n=5)')
plt.show()
```

# 6-Negatif Binom Dağılımı:


Negatif Binom Dağılımı:

* Başarıları saymak için kesikli bir olasılık dağılımıdır.
* Bir denemede başarı olasılığı p olan bir olayı düşünelim. r başarıya ulaşmak için kaç deneme yapmamız gerektiğini (başarısızlık sayısını) modellemek için kullanılır.
* Olasılık kütle fonksiyonu (PMF) şu şekildedir: P(X=n) = C(n-1, r-1) * p^r * q^(n-r)








```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Başarı sayısı (r)
r = 5

# Başarı olasılığı
p = 0.3

# X değerleri (deneme sayısı)
x = np.arange(r, r+10)

# Negatif Binom Dağılımı olasılık kütle fonksiyonu (PMF)
pmf = nbinom.pmf(x, r, p)
print(pmf)

# Grafik oluşturma
plt.bar(x, pmf)
plt.xticks(x)
plt.ylabel('Olasılık')
plt.xlabel('Deneme Sayısı')
plt.title('Negatif Binom Dağılımı (r=5, p=0.3)')
plt.show()
```

# 7-Multinomiyal Dağılım:


Multinomiyal Dağılım:

* Birden fazla kategoride birden fazla kez deney yapıldığında kullanılır.
* Örneğin, bir zarı üç kere atıp her sonuç için olasılıkları modellemek için kullanılabilir.
* Olasılık kütle fonksiyonu (PMF) şu şekildedir:
    P(X1=x1, X2=x2, ..., XK=xK) = (n! / (x1! * x2! * ... * xK!)) * (p1^x1) * (p2^x2) * ... * (pK^xK)






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multinomial

# Toplam deneme sayısı
n = 10

# Kategorilerin sayısı
K = 3

# Başarı sayıları
x = [2, 5, 3]

# Başarı olasılıkları
p = [0.3, 0.5, 0.2]

# Multinomiyal Dağılım olasılık kütle fonksiyonu (PMF)
pmf = multinomial.pmf(x, n, p)
print(pmf)

# Grafik oluşturma
categories = np.arange(1, K+1)
plt.bar(categories, pmf)
plt.xticks(categories)
plt.ylabel('Olasılık')
plt.xlabel('Kategoriler')
plt.title('Multinomiyal Dağılım (n=10, K=3)')
plt.show()

```

# Sürekli Olasılık Dağılımları


# 1-Normal Dağılım (Gauss Dağılımı):


Normal Dağılım (Gauss Dağılımı):
	
*	En yaygın kullanılan sürekli olasılık dağılımıdır.
*	Merkezi limit teoremi gereği pek çok doğal süreçte rastgele değişkenlerin dağılımları genellikle normal dağılıma yakınsar.
*	Ortalama (μ) ve standart sapma (σ) ile tanımlanır.
*	Tek zirveli ve simetrik bir dağılımdır.
*	Çan şeklinde bir grafik ve negatif sonsuza ve pozitif sonsuza doğru uzanan kuyruklara sahiptir.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Ortalama ve standart sapma
mu = 0
sigma = 1

# X değerleri
x = np.linspace(-5, 5, 1000)

# Normal Dağılım yoğunluk fonksiyonu (PDF)
pdf = norm.pdf(x, mu, sigma)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluğu')
plt.title('Normal Dağılım (mu=0, sigma=1)')
plt.grid(True)
plt.show()
```

# 2-Üniform Dağılım:


Üniform Dağılım:
	
*	Belirli bir aralıkta tüm değerlerin eşit olasılıkla gerçekleştiği dağılımdır.
*	A ve B gibi iki parametre ile tanımlanır ve değerler arasında homojen bir dağılım gösterir.
*	Tüm olasılık yoğunluk fonksiyonları (PDF) eşit olduğundan grafiği düz bir çizgidir.





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Minimum ve maksimum değerler
a = 0
b = 10

# X değerleri
x = np.linspace(a-2, b+2, 1000)

# Üniform Dağılım yoğunluk fonksiyonu (PDF)
pdf = uniform.pdf(x, loc=a, scale=b-a)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluğu')
plt.title('Üniform Dağılım (a=0, b=10)')
plt.grid(True)
plt.show()
```

# 3-Beta Dağılımı:

Beta Dağılımı:
	
*	Olasılık değerlerinin aralığını modellemek için kullanılır ve genellikle oranları temsil etmek için tercih edilir.
*	İki şekillendirme parametresi (α ve β) ile tanımlanır.
*	0 ile 1 arasındaki değerler için kullanılır ve çeşitli şekillere sahip olabilir.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Beta Dağılımı şekillendirme parametreleri
alpha = 2
beta_val = 5

# X değerleri
x = np.linspace(0, 1, 1000)

# Beta Dağılımı yoğunluk fonksiyonu (PDF)
pdf = beta.pdf(x, alpha, beta_val)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluğu')
plt.title('Beta Dağılımı (alpha=2, beta=5)')
plt.grid(True)
plt.show()

```

# 4-Gamma Dağılımı:

Gamma Dağılımı:
	
*	Pozitif sürekli değişkenlerin modellenmesi için kullanılır ve özellikle süreç zamanları gibi alanlarda sıkça kullanılır.
*	İki parametre (α ve β) ile tanımlanır.
*	Eksponeksiyon dağılımının genelleştirilmiş halidir.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Şekil parametresi (alpha)
alpha = 2

# Ölçek parametresi (beta)
beta = 2

# X değerleri (0'dan itibaren 10'a kadar)
x = np.linspace(0, 10, 100)

# Gamma Dağılımı olasılık yoğunluk fonksiyonu (PDF)
pdf = gamma.pdf(x, alpha, scale=beta)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Gamma Dağılımı (alpha=2, beta=2)')
plt.show()
```

# 5-Exponential Dağılımı:



Exponential Dağılımı:
	
*	İlk olayın gerçekleşmesi arasındaki sürenin modellemesinde kullanılır.
*	Yarıçaplanma süresi (λ) olarak bilinen tek bir parametresi vardır.
*	Belleksiz (memoryless) dağılım olarak bilinir, yani gelecekte bir olayın gerçekleşme süresi geçmiş olayların gerçekleşme süresine bağlı değildir.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Ölçek parametresi (lambda)
lambd = 0.5

# X değerleri (0'dan itibaren 5'e kadar)
x = np.linspace(0, 5, 100)

# Exponential Dağılımı olasılık yoğunluk fonksiyonu (PDF)
pdf = expon.pdf(x, scale=1/lambd)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Exponential Dağılımı (lambda=0.5)')
plt.show()
```

# 6-Weibull Dağılımı:



Weibull Dağılımı:
	
*	Ömrü modellemek için kullanılan bir dağılımdır ve özellikle güvenilirlik analizlerinde yaygın olarak kullanılır.
*	İki parametre (λ ve k) ile tanımlanır ve farklı şekillere sahip olabilir.







```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Şekil parametresi (k)
k = 2

# Ölçek parametresi (lambda)
lambda_w = 1

# X değerleri (0'dan itibaren 5'e kadar)
x = np.linspace(0, 5, 100)

# Weibull Dağılımı olasılık yoğunluk fonksiyonu (PDF)
pdf = weibull_min.pdf(x, k, scale=lambda_w)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Weibull Dağılımı (k=2, lambda=1)')
plt.show()
```

Pareto Dağılımı:

*	Pareto prensibine dayalı olarak zenginlerin daha zengin, yoksulların daha yoksul olma eğilimine işaret eder.





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Şekil parametresi (alpha)
alpha = 3

# Ölçek parametresi (xm)
xm = 1

# X değerleri (1'den itibaren 5'e kadar)
x = np.linspace(1, 5, 100)

# Pareto Dağılımı olasılık yoğunluk fonksiyonu (PDF)
pdf = pareto.pdf(x, alpha, scale=xm)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Pareto Dağılımı (alpha=3, xm=1)')
plt.show()
```

# 8-Log-Normal Dağılımı:



Log-Normal Dağılımı:
	
*	Logaritma dönüşümü sonrası normal dağılıma benzer, ancak asıl değişkenin üzerindeki pozitif değerler için kullanılır.
*	Finansal piyasalarda ve stok fiyatları gibi alanlarda sıkça kullanılır.





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Log-Normal Dağılım parametreleri (mu ve sigma)
mu = 1.0
sigma = 0.5

# X değerleri (oluşturulan değer aralığı)
x = np.linspace(0.01, 5, 100)

# Log-Normal Dağılım olasılık yoğunluk fonksiyonu (PDF)
pdf = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Log-Normal Dağılımı')
plt.show()
```

Cauchy Dağılımı:

# 10-Student-t Dağılımı:



Student-t Dağılımı:
	
*	Küçük örneklem büyüklüklerinde kullanılan ve normal dağılıma benzeyen bir dağılımdır.
*	Daha geniş kuyruklara sahiptir, bu nedenle normal dağılıma göre daha ağırdır.
*	T-testi ve diğer hipotez testlerinde kullanılır.






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Serbestlik derecesi (df)
df = 5

# X değerleri (oluşturulan değer aralığı)
x = np.linspace(-5, 5, 100)

# Student-t Dağılım olasılık yoğunluk fonksiyonu (PDF)
pdf = t.pdf(x, df=df)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('Student-t Dağılımı (df=5)')
plt.show()
```

# 11-F Dağılımı:




F Dağılımı:
	
*	İki varyansı karşılaştırmak için kullanılan bir dağılımdır.
*	Özellikle regresyon analizinde, ANOVA testlerinde ve varyans analizlerinde kullanılır.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Serbestlik dereceleri (dfn ve dfd)
dfn = 5
dfd = 10

# X değerleri (oluşturulan değer aralığı)
x = np.linspace(0.01, 5, 100)

# F Dağılım olasılık yoğunluk fonksiyonu (PDF)
pdf = f.pdf(x, dfn=dfn, dfd=dfd)
print(pdf)

# Grafik oluşturma
plt.plot(x, pdf)
plt.xlabel('Değerler')
plt.ylabel('Olasılık Yoğunluk Fonksiyonu (PDF)')
plt.title('F Dağılımı (dfn=5, dfd=10)')
plt.show()
```
